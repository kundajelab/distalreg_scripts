```python
%matplotlib inline
from scipy.stats import spearmanr,pearsonr
import matplotlib.pyplot as plt

def plot_proseq(y_valid_preds, y_valid): 
    print("Spearman R", spearmanr(y_valid_preds,y_valid))
    print("Pearson R", pearsonr(y_valid_preds,y_valid))
    plt.plot(y_valid_preds, y_valid, 'ro', markersize=2)
    plt.xlabel('Predictions')
    plt.ylabel('Actual')
    plt.title('Correlation graph')
    return plt

def evaluate_proseq(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = mean_squared_error(test_labels, predictions)
    plt = plot_proseq(predictions, test_labels)
    #mape = 100 * np.mean(errors / test_labels)
    #accuracy = 100 - mape
    print('Model Performance')
    print('MSError: {:0.4f} '.format(errors))
    #print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return predictions, _ #accuracy
```

```python
def plot_x(model, y_train_preds, y_train, y_valid_preds, y_valid, outdir, label, save=False):
    print("Train Spearman R:", spearmanr(y_train_preds,np.arcsinh(y_train)))
    print("Train Pearson R:", pearsonr(y_train_preds,np.arcsinh(y_train)))
    print("Valid Spearman R:", spearmanr(y_valid_preds,np.arcsinh(y_valid)))
    print("Valid Pearson R:", pearsonr(y_valid_preds,np.arcsinh(y_valid)))
    f, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
    ax1.plot(y_train_preds, np.arcsinh(y_train), 'ro', markersize=2)
    ax2.plot(y_valid_preds, np.arcsinh(y_valid), 'ro', markersize=2)
    ax1.set_xlabel('Train Predictions')
    ax1.set_ylabel('Observed Labels')
    ax2.set_xlabel('Valid Predictions')
    ax2.set_ylabel('Observed Labels')
    ax1.set_title('Correlation graph')
    
    def write_to_file():
        f.savefig(os.path.join(outdir, "{}_{}.pdf".format(str(model)[0], label)), format="pdf")
        with open(os.path.join(outdir, "{}_{}.txt".format(str(model)[0], label)), "w") as f:
            f.write("Train Spearman R:")
            f.write("\t")
            f.write(str(spearmanr(y_train_preds,np.arcsinh(y_train))))
            f.write("\n")
            f.write("Valid Spearman R:")
            f.write("\t")
            f.write(str(spearmanr(y_valid_preds,np.arcsinh(y_valid))))
            f.write("\n")
            f.write("Train Pearson R:")
            f.write("\t")
            f.write(str(pearsonr(y_train_preds,np.arcsinh(y_train))))
            f.write("\n")
            f.write("Valid Pearson R:")
            f.write("\t")
            f.write(str(pearsonr(y_valid_preds,np.arcsinh(y_valid))))
            f.close()
        return plt

    return plt
```

```python
def evaluate(model, train_features, train_labels, test_features, test_labels, outdir, label):
    predictions = model.predict(test_features)
    train_preds = model.predict(train_features)
    errors = mean_squared_error(np.arcsinh(test_labels), predictions)
    plt = plot_x(model, train_preds, train_labels, predictions, test_labels, outdir, label)
        
    print('Model Performance')
    print('MSError: {:0.4f} '.format(errors))
    return train_preds, predictions
```

### Boosted Linear Regression for deepsea embeddings

```python
boosted_linreg = LinearRegression()
y_labels = np.arcsinh(y_train_regression_labels)
y_predictions=0
y_residual=0
acc_weights={}
for i in range(20): 
    boosted_linreg.fit(x_train, y_labels)
    acc_weights[i]=boosted_linreg.coef_
    y_predictions += boosted_linreg.predict(x_train)
    y_residual = y_labels - y_predictions
    y_labels = y_residual
```

```python
boosted_fit = pd.DataFrame()
boosted_fit['y_predictions'] = y_predictions
boosted_fit['y_actual'] = np.arcsinh(y_train_regression_labels)

residual_fit = pd.DataFrame()
residual_fit['y_residual'] = y_residual
residual_fit['y_actual'] = np.arcsinh(y_train_regression_labels)

sns.lmplot(x='y_predictions', y='y_actual', data=boosted_fit)
print("Spearman R", scipy.stats.spearmanr(y_predictions,np.arcsinh(y_train_regression_labels)))
print("Pearson R", scipy.stats.pearsonr(y_predictions,np.arcsinh(y_train_regression_labels)))
```

```python
sns.lmplot(x='y_residual', y='y_actual', data=residual_fit)
print("Spearman R", scipy.stats.spearmanr(y_residual,np.arcsinh(y_train_regression_labels)))
print("Pearson R", scipy.stats.pearsonr(y_residual,np.arcsinh(y_train_regression_labels)))
```
